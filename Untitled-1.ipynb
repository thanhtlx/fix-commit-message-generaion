```python
print(df.shape)
df = df.drop_duplicates(subset=['commit_id','old_path_file','new_path_file'])
print(df.shape)
```

```python
from collections import defaultdict
diffs = defaultdict(lambda:set())
messages = dict()
for _,row in df.iterrows():
    # print(row)
    cm = row['commit_id']
    if len(cm) != 40:
        continue
    if not isinstance(row['diff'],str):
        continue
    diffs[cm].add(row['diff'])
    if isinstance(row['label'],str):
        messages[cm] = row['label'].splitlines()[0]
    else:
        print(cm,row['label'])
        # messages[cm] = ''
```

```python
res = list()
for k,v in messages.items():
    # print(diffs[k])
    res.append({'commit_id':k, 'diff': '\n'.join(list(diffs[k])),'message':v})
print(len(res))
```

```python
ds = pd.DataFrame(res)
ds.head(2)['message'].values[1]
```

```python
ds.to_csv('diff_msg.csv')
```

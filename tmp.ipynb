```python
import json 
with open('db_cve.json') as f:
    data = json.load(f)
```

```python
keys = list(data.keys())
keys[:5]
```

```python
len(keys), len(set(keys))
```

```python
with open('repo.txt') as f:
    repos = [l.strip() for l in f.readlines()]
```

```python
name_to_link= dict()
name_to_cm = dict()
print(len(keys))
for k in keys:
    el = k.split('/')
    while len(el) > 0 and ('.com' not in el[0] and '.org' not in el[0]):
        el.pop(0)
    el = el[1:]
    if len(el) < 2:
        print("??")
        continue
    rp_name = "___".join(el)
    rp_name = rp_name.lower()
    if rp_name not in repos:
        print('not in',rp_name)
        continue
    if rp_name in name_to_link:
        if 'github' in k:
            name_to_link[rp_name] = k 
    else:
        name_to_link[rp_name] = k 
    if rp_name in name_to_cm:
        name_to_cm[rp_name] = name_to_cm[rp_name] + data[k]
    else:
        name_to_cm[rp_name] = data[k]
        # pass
    # else:
        # print(k)
len(name_to_link), len(name_to_cm)
```

```python
with open('name_to_cm.json','w+') as f:
    json.dump(name_to_cm,f)
with open('name_to_link.json','w+') as f:
    json.dump(name_to_link,f)
```

```python
for k,v in name_to_link.items():
    cmd = f'git clone {v} {k}'
    print(cmd)
```

```python
result = list()

for k,v in name_to_cm.items():
    for cm in v:
        tmp = {'repo_name': k,'fixing_commit':{'commit_hash':cm,'commit_timestamp':'0'},'vulnerability_contributing_commits':[]}
        result.append(json.dumps(tmp))
    # print(k,len(v))
    # break
len(result)
```

```python
with open('extend.json','w+') as f:
    f.write('\n'.join(result))
```
